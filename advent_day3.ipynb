```python
f = open("drive/MyDrive/Colab Notebooks/advent_day3.txt", "r")
file_contents = f.read()
f.close()
```

```python
# prompt: create a regex of the expression mul(2,2) where the 2s can be any set of numbers

import re

regex = r"mul\(\d+,\d+\)"

# Example usage
strings = [
    "mul( 2,2)",
    "mul(10,20)",
    "mul(1,1)",
    "mul(2,222)",
    "add(2,2)",
    "mul(a,b)"
]

for string in strings:
  if re.match(regex, string):
    print(f"'{string}' matches the regex")
  else:
    print(f"'{string}' does not match the regex")
```

```python
regex = r"mul\(\d+,\d+\)"
matches = re.findall(regex, file_contents)
print(matches)
```

```python
sum = 0
for match in matches:
  numbers = re.findall(r'\d+', match)
  sum += int(numbers[0])*int(numbers[1])

print(sum)
```

```python
groups = re.findall(r"(mul\(\d+,\d+\))|(do\(\))|(don't\(\))", file_contents)
expressions = [x for row in groups for x in row if x != ""]
expressions
```

```python
enabled = True
sum = 0
for expression in expressions:
  if expression == "don't()":
    enabled = False
  elif expression == "do()":
    enabled = True
  elif enabled:
    numbers = re.findall(r'\d+', expression)
    sum += int(numbers[0])*int(numbers[1])
sum
```
